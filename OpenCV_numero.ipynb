```python
import cv2
import imutils
import numpy as np
from keras.models import load_model
```

```python
# Habilitar camara para video
CAMERA = cv2.VideoCapture(0)

#definicion de marco de deteccion de numeros
CAPTURE_WIDTH = 400
ROI_LONG = 150 # Region de analisis
MARGIN = 50
TOP = MARGIN
RIGHT = CAPTURE_WIDTH - MARGIN
BOTTOM = TOP + ROI_LONG
LEFT = RIGHT - ROI_LONG
```

```python
#cargar el modelo, la ruta debe ser modificada por la del sitio donde se creo el archivo .h5
model = load_model("C:/Users.../CNN_numeros.h5")
```

```python
while(True):
    _, frame = CAMERA.read() #comenzar el video
    frame = imutils.resize(frame, CAPTURE_WIDTH)
    (height, width) = frame.shape[:2]

    # agregar un cuadro en la region de interes
    cv2.rectangle(frame, (LEFT, TOP), (RIGHT, BOTTOM), (0,255,0), 2)

    # preprocesamiento en la region de interes 
    roi = frame[TOP+2:BOTTOM-2, LEFT+2:RIGHT-2]
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY) #convierte la imagen de la region de interes en b/n
    _, gray = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY_INV) # Se invierte para mejor diferenciacion del numero
    cv2.imshow("ROI", gray)

    # Predict and show prediction
    gray_small = imutils.resize(gray, 28)
    gray_small = gray_small.reshape(1,28,28,1)
    pred = model.predict(gray_small)[0]
    classes_x=np.argmax(pred)
    LABEL_TEXT = str(classes_x)
    LABEL_COLOR = (0,255,0)
    cv2.putText(frame, LABEL_TEXT, (LEFT, TOP-7), cv2.FONT_HERSHEY_SIMPLEX, 1, LABEL_COLOR, 2)
    cv2.imshow("Frame", frame)
    #la prediccion se muestra directamente en el video
    #adicionalmente se abre la ventana de la region de interes 

    # "q" para salir del While, cerrar video y parar reconocimiento
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cv2.destroyAllWindows()
```
